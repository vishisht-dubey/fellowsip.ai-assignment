```python
import nltk
```

```python
nltk.download('all')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer
```

```python
def preprocessing_text(data):
    word_tokens = word_tokenize(data.lower())
    filtered_tokens = [token for token in word_tokens if token not in stopwords.words('english')]
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text  
```

```python
# print(preprocessing_text("Vishisht is a nice guy with a decent coding skills."))
```

```python
df['review'] = df['review'].apply(preprocessing_text)
```

```python
df
```

```python
def transform_response(type):
    if(type=='positive'):
        return 1
    return 0
```

```python
df['sentiment'] = df['sentiment'].apply(transform_response)
```

```python
df
```

```python
analyzer = SentimentIntensityAnalyzer()
```

```python
def get_sentiment(text):

    scores = analyzer.polarity_scores(text)
    print(scores)
    sentiment = 1 if scores['pos'] > scores['neg'] and scores['compound']>=0 else 0
    # sentiment = 1 if scores['compound'] >=0 else 0

    return sentiment
```

```python
df['generated_sentiment'] = df['review'].apply(get_sentiment)
```

```python
df
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(df['sentiment'],df['generated_sentiment']))
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['sentiment'],df['generated_sentiment']))
```
